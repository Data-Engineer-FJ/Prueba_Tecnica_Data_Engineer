```python
import sqlite3
conn = sqlite3.connect('test.db')

#### PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
    SELECT
        Equipo,
        grupo AS Grupo,
        SUM(puntos) AS Puntos,
        SUM(dif) AS Gol_Diferencia
    FROM (
        -- Analizamos al equipo cuando es LOCAL
        SELECT
            TRIM(SUBSTR(partido, 1, INSTR(partido, ' vs. ') - 1)) AS Equipo,
            grupo,
            CASE
                WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) > CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) THEN 3
                WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) = CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) THEN 1
                ELSE 0
            END AS puntos,
            CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) - CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) AS dif
        FROM Partidos

        UNION ALL

        -- Analizamos al equipo cuando es VISITANTE
        SELECT
            TRIM(SUBSTR(partido, INSTR(partido, ' vs. ') + 5)) AS Equipo,
            grupo,
            CASE
                WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) > CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) THEN 3
                WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) = CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) THEN 1
                ELSE 0
            END AS puntos,
            CAST(SUBSTR(resultado, INSTR(resultado, ',') + 1) AS INT) - CAST(SUBSTR(resultado, 1, INSTR(resultado, ',') - 1) AS INT) AS dif
        FROM Partidos
    )
    GROUP BY Equipo, grupo
    ORDER BY Puntos DESC, Gol_Diferencia DESC;
''')

# Imprimimos el resultado con formato de tabla
print(f"{'EQUIPO':<30} | {'GRUPO':<10} | {'PTS':<5} | {'GD'}")
print("-" * 55)
for row in cursor:
    # Quitamos puntos finales que vienen en la data (ej: 'Espana.')
    nombre_limpio = row[0].strip('.')
    print(f"{nombre_limpio:<30} | {row[1]:<10} | {row[2]:^5} | {row[3]:^3}")

conn.close()
```

```python
# Ejercicio # 2
conn = sqlite3.connect('test.db')

#### PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
    WITH Tabla_Posiciones AS (
        SELECT
            Equipo,
            grupo,
            SUM(puntos) AS PTS,
            SUM(dif) AS GD
        FROM (
            SELECT TRIM(SUBSTR(partido, 1, INSTR(partido, ' vs.') - 1)) AS Equipo, grupo,
                   CASE WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) > CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) THEN 3
                        WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) = CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) THEN 1 ELSE 0 END AS puntos,
                   CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) - CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) AS dif
            FROM Partidos
            UNION ALL
            SELECT TRIM(SUBSTR(partido, INSTR(partido, 'vs. ') + 4)) AS Equipo, grupo,
                   CASE WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) > CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) THEN 3
                        WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) = CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) THEN 1 ELSE 0 END AS puntos,
                   CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) - CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) AS dif
            FROM Partidos
        )
        GROUP BY Equipo, grupo
    ),
    Ranking_Clasificados AS (
        SELECT *,
               RANK() OVER (PARTITION BY grupo ORDER BY PTS DESC, GD DESC) as posicion
        FROM Tabla_Posiciones
    )
    -- Seleccionamos solo los 2 mejores de cada grupo
    SELECT * FROM Ranking_Clasificados WHERE posicion <= 2;
''')

print(f"{'EQUIPO':<25} | {'GRUPO':<10} | {'POS':<4} | {'PTS':<4} | {'GD'}")
print("-" * 55)
for row in cursor:
    print(f"{row[0].replace('.', ''):<25} | {row[1]:<10} | {row[4]:^4} | {row[2]:^4} | {row[3]:^3}")

conn.close()
```

```python
# Ejercicio # 3
conn = sqlite3.connect('test.db')

#### PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''
    WITH Posiciones AS (
        SELECT Equipo, grupo,
               RANK() OVER (PARTITION BY grupo ORDER BY SUM(puntos) DESC, SUM(dif) DESC) as rank
        FROM (
            SELECT TRIM(SUBSTR(partido, 1, INSTR(partido, ' vs.') - 1)) AS Equipo, grupo,
                   CASE WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) > CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) THEN 3
                        WHEN CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) = CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) THEN 1 ELSE 0 END AS puntos,
                   CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) - CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) AS dif
            FROM Partidos
            UNION ALL
            SELECT TRIM(SUBSTR(partido, INSTR(partido, 'vs. ') + 4)) AS Equipo, grupo,
                   CASE WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) > CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) THEN 3
                        WHEN CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) = CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) THEN 1 ELSE 0 END AS puntos,
                   CAST(SUBSTR(resultado, INSTR(resultado, ',')+1) AS INT) - CAST(SUBSTR(resultado, 1, INSTR(resultado, ',')) AS INT) AS dif
            FROM Partidos
        )
        GROUP BY Equipo, grupo
    ),
    Equipos_Finales AS (
        -- Creamos una descripción que coincida con la tabla Clasificados
        -- Ej: '1º del grupo A'
        SELECT Equipo, (rank || 'º del ' || LOWER(grupo)) as desc_posicion
        FROM Posiciones
        WHERE rank <= 2
    )
    -- Cruzamos los resultados reales con la tabla de Sedes y Horarios
    SELECT
        c.partido AS Llave_Original,
        e1.Equipo AS Local_Real,
        e2.Equipo AS Visitante_Real,
        c.fecha,
        c.sede
    FROM Clasificados c
    JOIN Equipos_Finales e1 ON c.partido LIKE e1.desc_posicion || '%'
    JOIN Equipos_Finales e2 ON c.partido LIKE '%' || e2.desc_posicion
    ORDER BY c.fecha ASC;
''')

print(f"{'LLAVE':<40} | {'ENCUENTRO REAL':<35} | {'SEDE'}")
print("-" * 100)
for row in cursor:
    llave = row[0]
    encuentro = f"{row[1].replace('.','')} vs {row[2].replace('.','')}"
    sede = row[4]
    print(f"{llave:<40} | {encuentro:<35} | {sede}")

conn.close()
```

```python
from typing import cast
import sys
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import explode, col, desc, avg, min, max, rank, asc, round, pow, sqrt
from pyspark.sql import Window
from pyspark.sql.functions import udf

# Configuración de la sesión
spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

####### PROGRAME SU RESPUESTA AQUI ##############

path = "CENTROS_EDUCATIVOS_MADRID.json"

try:
    # 5.1 Leer el archivo con multiLine (esto ya funciona por lo que veo)
    df = spark.read.option("multiLine", True).json(path)

    # Aseguramos que las coordenadas sean números para los cálculos
    df = df.withColumn("direccion_coor_x", col("direccion_coor_x").cast("double")) \
           .withColumn("direccion_coor_y", col("direccion_coor_y").cast("double"))

    # 5.2.1 Determinar el promedio por grupo "centro_titularidad"
    window_tit = Window.partitionBy("centro_titularidad")

    df_promedios = df.withColumn("avg_x", avg("direccion_coor_x").over(window_tit)) \
                     .withColumn("avg_y", avg("direccion_coor_y").over(window_tit))

    # 5.2.2 UDF para calcular la distancia euclideana
    def calcular_distancia(x, y, ax, ay):
        if x is None or y is None or ax is None or ay is None: return 0.0
        return float(((x - ax)**2 + (y - ay)**2)**0.5)

    distancia_udf = udf(calcular_distancia, DoubleType())

    df_con_dist = df_promedios.withColumn(
        "distancia_euclideana",
        distancia_udf(col("direccion_coor_x"), col("direccion_coor_y"), col("avg_x"), col("avg_y"))
    )

    # 5.2.3 Identificar el centro con menor distancia por grupo
    # Ordenamos por la nueva columna "distancia_euclideana"
    window_min = Window.partitionBy("centro_titularidad").orderBy(asc("distancia_euclideana"))

    # Filtramos para quedarnos con el más cercano (rank 1)
    # CORRECCIÓN: usamos 'centro_nombre' en lugar de 'nombre_centro'
    resultado = df_con_dist.withColumn("posicion", rank().over(window_min)) \
                           .filter(col("posicion") == 1) \
                           .drop("posicion")

    # Mostrar resultados finales con los nombres de columna correctos
    resultado.select(
        "centro_titularidad",
        "centro_nombre",
        "distancia_euclideana"
    ).show()

except Exception as e:
    print(f"Error detectado: {e}")

#################################################
```
